```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dir = "/Users/omard/Downloads/SCRATCHFOLDER/"
qe = np.loadtxt(dir+"QEwebsky.txt")
qedel = np.loadtxt(dir+"QEdelwebsky.txt")


bin_edges = np.arange(20, 3000, 100)
cents = (bin_edges[1:] + bin_edges[:-1]) / 2
```

```python
qe_key_ = "ptt"
term = "PB"
llplot, interpolated = np.loadtxt(f"n32_{qe_key_}_{0}_{term}_GM.txt", unpack = True)

llplot, interpolated1 = np.loadtxt(f"n32_{qe_key_}_{1}_{term}_GM.txt", unpack = True)
```

```python
plt.title("Websky sims with Giulio's theory")
plt.plot(llplot, interpolated, label = "Theory QE", color = "blue")
plt.plot(llplot, interpolated1, color = "red", label = "Theory QE Delensed")

plt.plot(llplot, llplot*0., ls = '--', color = "black")

plt.plot(cents, qe*100, label = "QE", color = "blue", alpha = 0.4)
plt.plot(cents, qedel*100, label = "QE from Delensed Maps with $W\phi$", alpha = 0.4, color = "red")
plt.ylim(-4, 1)
plt.xlabel("L")
plt.ylabel("N32")
plt.legend()
```
