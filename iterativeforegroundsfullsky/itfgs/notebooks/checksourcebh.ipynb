```python
import matplotlib.pyplot as plt
import numpy as np

import healpy as hp

from plancklens import utils
```

```python
copy = lambda alm: utils.alm_copy(alm, lmax = 4500)
```

```python
path = "/Users/omard/Downloads/SCRATCHFOLDER/n32/S4WebskyWebskyBorn/lenscarfrecs/ptt_bh_s_sim0000bh/"

sfile = "normalized_s_slm_it000.npy"

pfile = "normalized_phi_plm_it000.npy"

inputphi = "/Users/omard/Downloads/SCRATCHFOLDER/n32/S4WebskyWebskyBorn/cmbs/plm_in_0_lmax5120.fits"
```

```python
s = np.load(path + sfile)
p = np.load(path + pfile)
phiinp = hp.read_alm(inputphi)
phiinp = copy(phiinp)
```

```python
len(s), len(p), len(phiinp)
```

```python
cls = hp.alm2cl(s, phiinp)
clp = hp.alm2cl(p, phiinp)
clinp = hp.alm2cl(phiinp, phiinp)
```

```python
plt.title("Autos")
plt.loglog(abs(cls), label = "s")
plt.loglog(clp, label = "p")
plt.loglog(clinp, label = "input")
plt.legend()
```

```python
#now plot crosses with input

clppi = hp.alm2cl(p, phiinp)
clspi = hp.alm2cl(s, phiinp)

plt.figure()
plt.title("Crosses")
plt.plot(clppi, label = "p x input")
plt.loglog(abs(clspi), label = "abs s x input")
plt.legend()
```

```python
#now plot crosses with p

clpp = hp.alm2cl(p, p)
clsp = hp.alm2cl(s, p)

plt.figure()
plt.title("Crosses")
plt.plot(clpp, label = "p x p")
plt.plot(abs(clsp), label = "abs s x p")
plt.yscale("log")
plt.legend()
```
